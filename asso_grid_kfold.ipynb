```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
df = pd.read_csv('asso.csv')
df.head()
```

```python
model= apriori(df, min_support=0.07, use_colnames=True)
rules = association_rules(model, metric="lift", min_threshold=2)
rules.head()
```

```python
rules[ (rules['lift'] >= 2) &
      (rules['confidence'] >= 0.8) ]
```

```python
data=pd.read_csv('admission.csv')
data.head(1)
x=data.iloc[:,[1,2,3,6,7]]
y=data.iloc[:,8]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn import metrics
param_grid = {'C': [0.1,1, 10, 100], 'gamma': [1,0.1,0.01,0.001]} 
grid = GridSearchCV(SVC(),param_grid)
grid.fit(x,y)
```

```python
grid.get_params
```

```python
y_pred=grid.predict(x)
print(metrics.accuracy_score(y,y_pred))
```

```python
from sklearn import model_selection
x=data.iloc[:,1:8]
y=data.iloc[:,8]
x_train,x_test,y_train,y_test=model_selection.train_test_split(x,y,test_size=0.3,random_state=1)
k_fold=model_selection.KFold(n_splits=5,random_state=1)
from sklearn import tree
model=tree.DecisionTreeClassifier()
results=model_selection.cross_val_score(model,x_train,y_train,cv=k_fold)
print(results)
```
